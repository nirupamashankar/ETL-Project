```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "winemag.csv"
winemag_df = pd.read_csv(csv_file)
```

```python
csv_file2 = "white_wine.csv"
white_df = pd.read_csv(csv_file2)
```

```python
csv_file3 = "majestic.csv"
majestic_df = pd.read_csv(csv_file3)
```

```python
# get only the columns needed
winemag_df = winemag_df[['country', 'description', 'price', 'province', 'title', 
                         'variety', 'winery']]

# renaming columns, making them consistent across the dbs
winemag_df = winemag_df.rename(columns={"country": "Country",
                                        "description": "Description",
                                        "price": "Price",
                                        "province": "Region",
                                        "title": "Full Name",
                                        "variety": "Wine Type",
                                        "winery": "Winery"})
# adding source column
winemag_df["Source"] = 'Winemag'

# reordering the columns to put Full Name first, as similar as possible with other dfs
winemag_df = winemag_df[['Full Name', 'Wine Type', 'Price', 'Country', 'Region', 
                         'Winery', 'Description', 'Source']]
winemag_df
```

```python
white_df = white_df[['FullName', 'Winery', 'Year', 'Region', 
                    'RegionalVariety', 'WineRating', 'WinePrice']]
white_df = white_df.rename(columns={"FullName": "Full Name",
                                    "Winery": "Winery",
                                    "Year": "Year",
                                    "Region": "Region",
                                    "RegionalVariety": "Wine Type",
                                    "WineRating": "Wine Rating",
                                    "WinePrice": "Price"})
white_df["Source"] = 'White Wine'
white_df = white_df[['Full Name', 'Wine Type', 'Price', 'Region',
                     'Winery', 'Year', 'Wine Rating', 'Source']]
white_df
```

```python
majestic_df = majestic_df[['Unnamed: 0', 'price', 'wine_type', 'country', 'rating_score']]
majestic_df = majestic_df.rename(columns={'Unnamed: 0': 'Full Name',
                                        'price': 'Price',
                                        'wine_type': 'Wine Type',
                                        'country': 'Country',
                                        'rating_score': 'Rating Score'})
majestic_df["Source"] = 'Majestic'
majestic_df = majestic_df[['Full Name', 'Wine Type', 'Price', 'Country', 
                           'Rating Score', 'Source']]
majestic_df 
```

```python
df1 = pd.concat([winemag_df, white_df])
df2 = pd.concat([df1, majestic_df])
df2.shape
```
