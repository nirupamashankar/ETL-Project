```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "winemag.csv"
winemag_df = pd.read_csv(csv_file)
winemag_df.head() 
```

```python
csv_file2 = "white_wine.csv"
white_wine_df = pd.read_csv(csv_file2)
white_wine_df.head() 
```

```python
csv_file3 = "majestic.csv"
majestic_df = pd.read_csv(csv_file3)
majestic_df.head()
```

```python
csv_file4 = "QualityReds.csv"
quality_reds_df = pd.read_csv(csv_file4)
quality_reds_df.head()
```

```python
csv_file5 = "QualityWhites.csv"
quality_whites_df = pd.read_csv(csv_file5)
quality_whites_df.head()
```

```python
# get only the columns needed
new_winemag_df = winemag_df[['country', 'description', 'price', 'province', 'title', 'variety', 'winery']]
new_winemag_df
# renaming columns, making them consistent across the dbs
new_winemag_df = new_winemag_df.rename(columns={"country": "Country",
                                               "description": "Description",
                                               "price": "Price",
                                               "province": "Region",
                                               "title": "Full Name",
                                               "variety": "Variety",
                                               "winery": "Winery"})
new_winemag_df
```

```python
new_white_wine_df = white_wine_df[['FullName', 'Winery', 'WineName', 'Year', 'Region', 
                                   'RegionalVariety', 'VintageRating', 'WineRating', 'WinePrice']]
new_white_wine_df = new_white_wine_df.rename(columns={"FullName": "Full Name",
                                                      "Winery": "Winery",
                                                      "WineName": "Wine Name",
                                                      "Year": "Year",
                                                      "Region": "Region",
                                                      "RegionalVariety": "Variety",
                                                      "VintageRating": "Vintage Rating",
                                                      "WineRating": "Wine Rating",
                                                      "WinePrice": "Price"})
new_white_wine_df
```

```python
new_majestic_df = majestic_df[['Unnamed: 0', 'price', 'wine_type', 'country', 'grape', 'region', 'rating_score']]
new_majestic_df = new_majestic_df.rename(columns={'Unnamed: 0': 'Full Name',
                                                 'price': 'Price',
                                                 'wine_type': 'Wine Type',
                                                 'country': 'Country',
                                                 'grape': 'Grape',
                                                  'region': 'Region',
                                                 'rating_score': 'Rating Score'})
new_majestic_df
```

```python
new_quality_reds_df = quality_reds_df[['Unnamed: 0', 'fixed.acidity', 'volatile.acidity', 
                                       'citric.acid', 'residual.sugar', 'chlorides',
                                       'free.sulfur.dioxide', 'total.sulfur.dioxide',
                                       'density', 'pH', 'sulphates', 'alcohol', 'quality']]
new_quality_reds_df = new_quality_reds_df.rename(columns={'Unnamed: 0': 'ID',
                                                         'fixed.acidity': 'Fixed Acidity',
                                                         'volatile.acidity': 'Volatile Acidity',
                                                         'citric.acid': 'Citric Acid',
                                                        'residual.sugar': 'Residual Sugar',
                                                        'chlorides': 'Chlorides',
                                                         'free.sulfur.dioxide': 'Free Sulfur Dioxide',
                                                          'total.sulfur.dioxide': 'Total Sulfur Dioxide',
                                                          'density': 'Density',
                                                          'pH': 'pH',
                                                         'sulphates': 'Sulphates',
                                                          'alcohol': 'Alcohol',
                                                          'quality': 'Quality'})
new_quality_reds_df
```

```python
new_quality_whites_df = quality_whites_df[['Unnamed: 0', 'fixed.acidity', 'volatile.acidity', 
                                       'citric.acid', 'residual.sugar', 'chlorides',
                                       'free.sulfur.dioxide', 'total.sulfur.dioxide',
                                       'density', 'pH', 'sulphates', 'alcohol', 'quality']]
new_quality_whites_df = new_quality_whites_df.rename(columns={'Unnamed: 0': 'ID',
                                                         'fixed.acidity': 'Fixed Acidity',
                                                         'volatile.acidity': 'Volatile Acidity',
                                                         'citric.acid': 'Citric Acid',
                                                        'residual.sugar': 'Residual Sugar',
                                                        'chlorides': 'Chlorides',
                                                         'free.sulfur.dioxide': 'Free Sulfur Dioxide',
                                                          'total.sulfur.dioxide': 'Total Sulfur Dioxide',
                                                          'density': 'Density',
                                                          'pH': 'pH',
                                                         'sulphates': 'Sulphates',
                                                          'alcohol': 'Alcohol',
                                                          'quality': 'Quality'})
new_quality_whites_df
```
